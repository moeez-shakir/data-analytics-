#Using Class

```python
class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, experience_years):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.experience_years = experience_years

    def describe_employee(self):
        print(f"Employee Name: {self.name}")
        print(f"Employee Number: {self.number}")
        print(f"Address: {self.address}")
        print(f"Salary: {self.salary}")
        print(f"Post: {self.post}")
        print(f"Year of Joining: {self.year_of_joining}")
        print(f"Education: {self.education}")
        print(f"Years of Experience: {self.experience_years}")
        print("-----------")

    def update_year_of_experience(self):
        self.experience_years += 1

    def increase_salary(self, increment):
        self.salary += increment


# Creating an instance called employee101
employee101 = Employee("John Doe", 101, "123 Main St, City", 50000, "Manager", 2015, "MBA", 8)

# Printing individual attributes
print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)

# Calling methods for employee101
employee101.describe_employee()
employee101.increase_salary(5000)  # Increase salary by $5000
employee101.update_year_of_experience()

# After updates
employee101.describe_employee()

# Creating 10 different instances
employees = [
    Employee("Alice Smith", 102, "456 Oak St", 60000, "Engineer", 2018, "BS in Engineering", 5),
    Employee("Bob Johnson", 103, "789 Pine St", 55000, "Accountant", 2017, "CPA", 6),
    Employee("Cathy Lee", 104, "101 Maple St", 48000, "Designer", 2020, "BFA", 3),
    Employee("David Brown", 105, "202 Birch St", 53000, "Marketing", 2019, "MBA", 4),
    Employee("Eve Miller", 106, "303 Cedar St", 59000, "Consultant", 2016, "PhD", 7),
    Employee("Frank Wilson", 107, "404 Elm St", 62000, "Analyst", 2015, "BS in Finance", 9),
    Employee("Grace Evans", 108, "505 Willow St", 50000, "HR", 2018, "MA in HR", 5),
    Employee("Hank Harris", 109, "606 Aspen St", 48000, "IT Support", 2020, "BS in IT", 3),
    Employee("Ivy Lewis", 110, "707 Cypress St", 57000, "Project Manager", 2016, "MBA", 8),
    Employee("Jake Clark", 111, "808 Redwood St", 55000, "Sales", 2017, "BA in Marketing", 6)
]

# Calling describe_employee() for each instance
for emp in employees:
    emp.describe_employee()

# Updating some instances
employees[0].increase_salary(2000)
employees[1].update_year_of_experience()

# Describing updated employees
employees[0].describe_employee()
employees[1].describe_employee()
```
