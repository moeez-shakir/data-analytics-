## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Replace income values using numpy.where
df['income'] = np.where(df['income'] <= 50000, 0, 1)

# Display the updated DataFrame
print("\nUpdated DataFrame:")
print(df)
```

## ***18. Which Workclass getting The Highest Salary***

```python
df
```

```python
print("DataFrame:")
print(df)

# Group by workclass and calculate the average income
average_salary_by_workclass = df.groupby('workclass')['income'].mean().reset_index()

# Find the workclass with the highest average salary
highest_salary_workclass = average_salary_by_workclass.loc[average_salary_by_workclass['income'].idxmax()]

# Display the result
print("\nWorkclass with the highest average salary:")
print(highest_salary_workclass)
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
df = pd.DataFrame(df)

# Convert 'income' column to numeric, handling non-numeric entries
df['income'] = pd.to_numeric(df['income'], errors='coerce')

# Calculate the number of persons with income > 50k by gender
income_above_50k = df[df['income'] > 50000].groupby('gender').size()

# Calculate the total number of persons by gender
total_by_gender = df.groupby('gender').size()

# Calculate the probability of income > 50k by gender
probability = (income_above_50k / total_by_gender).fillna(0)

# Display the results
print("\nProbability of getting salary > 50k by Gender:")
print(probability)

# Determine who has a better chance
better_chance = probability.idxmax()
print(f"\nThe gender with a better chance to earn >50k is: {better_chance}")
```

## ***20. Convert Workclass Column DataType To Category Datatype***

```python
df = pd.DataFrame(df)

# Display the original DataFrame and its dtypes
print("Original DataFrame:")
print(df)
print("\nData Types Before Conversion:")
print(df.dtypes)

# Convert 'workclass' column to category datatype
df['workclass'] = df['workclass'].astype('category')

# Display the updated DataFrame and its dtypes
print("\nUpdated DataFrame:")
print(df)
print("\nData Types After Conversion:")
print(df.dtypes)
```
